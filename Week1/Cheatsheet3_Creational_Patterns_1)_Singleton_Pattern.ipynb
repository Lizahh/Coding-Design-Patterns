# **A). Introduction:**

* Singleton is a creational pattern which describes a way to create an object (single or multiple??)

* The singleton pattern is one of the simplest examples of a design pattern. However, you will see that it is a powerful technique.

* As its name suggests, the singleton design pattern is about something singular, having only one. But what is it referring to? What do we only want one of? The Singleton pattern refers to having only one object of a class.

# **B). Example:**

Imagine you’re making a poker game app, and the player can change settings like:

1. Card design
2. Table color

These settings are saved in a Preferences object.

**Problem Without Singleton:**
If your app accidentally creates multiple Preferences objects:

1. One object might save red cards, and another might save blue cards.
2. The game won’t know which one to use.
3. This causes confusion and errors.

**Solution: Singleton Pattern:**

1. The Singleton makes sure there is only one Preferences object in the entire app.
2. No matter where in the app you access it, you always get the same object.
3. This keeps settings consistent and prevents conflicts.

**Why is this important?**
The game always uses the same settings, and the user sees the correct colors and designs every time.



# **C). How to Actually Make a Singleton?**

* We now know that a Singleton means having only one object of a class. But how do we make sure this actually happens?

## **a. The "Informal" Way (Bad Idea)**
* You create one object and tell yourself not to create another.
* But what if you forget?
* What if another developer on your team creates a second object by mistake?
* This can lead to bugs and confusion.

## **b. The Proper Way (Good Idea)**
* We write code in the class that prevents creating more than one object.
* Even if someone tries to make another object, the code won’t allow it.
* This way, the rule is built into the program, not just in your memory.

**Why is this important?**
In big projects with multiple developers, we need a foolproof way to enforce the Singleton rule so mistakes don’t happen.

# **D). Implementation:**

## **Step-by-Step Explanation:**

**1. Private Constructor:**

* Normally, a class has a **public constructor**, which means anyone can create an object of that class at any time.
* To prevent this, make the constructor **private**. This way, the constructor can't be called outside the class.

**2. Private Class Variable (uniqueInstance):**

* You need a variable that holds the single instance of your class.
* This variable is private so that no one can change it outside the class.
Let’s call it uniqueInstance.

**3. Public Method (getInstance):**

* Now, you need a public method that will create and return the instance of your class.
* We call this method getInstance.

### **Here’s how it works:**
1. Check if uniqueInstance is null. If it is, that means the object doesn't exist yet.
2. If it is null, the getInstance method creates the object and sets uniqueInstance to reference it.
3. If uniqueInstance is not null (meaning the object already exists), it returns the existing object.

## **Lazy Creation:**

One of the benefits of this method is lazy creation: the object is not created until it is actually needed. This is more efficient, especially if the object is large.

## **Thread Safety Issues:**

If multiple threads are trying to access the object at the same time, there could be problems. This needs special attention when using the Singleton pattern in a multi-threaded environment.

# **Why Use Singleton?**

Imagine you're making an app where the user preferences need to be saved. You only want one preferences object, because if there were more than one, they might cause confusion or inconsistencies in your app. The Singleton pattern ensures you only have one instance of the preferences object.

# **Advantages of Singleton:**

**Global Access:** The Singleton ensures the class is accessible from anywhere in the program.
**Lazy Creation:** The object is created only when it’s needed, which saves resources.
